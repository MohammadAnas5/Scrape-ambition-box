```python
# Apply for all pages 333 pages 

final = pd.DataFrame()
for j in range(1,333):
    
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    
    headers = {'User-Agent':"Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Mobile Safari/537.36"}
    webpage = requests.get(url,headers=headers).text  
    
    containers = soup.find_all('div',class_="company-content-wrapper")  
    
    companies_names = []
    star_rating = []
    reviews = [] 
    office_locations = []
    about = []


    for i in containers:
        companies_names.append(i.find('h2').text.strip()) 
        star_rating.append(i.find('p',class_='rating').text.strip())
        reviews.append(i.find('a',class_='review-count sbold-Labels').text.strip()) 
        office_locations.append(i.find('p',class_ = 'infoEntity').text.strip() ) 

    
    d = {'companies_names':companies_names,'star_rating':star_rating, 'reviews':reviews,'office_locations':office_locations}
    df = pd.DataFrame(d) # One page
    final=final.append(df, ignore_index = True) #all pages  
```
